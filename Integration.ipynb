## 1. Setup Code

The top portion of code covers basic imports from the matplotlib and numpy modules.

The bottom portion is of a function that will be the integrand.

```python
from matplotlib import pyplot as plt
import numpy as np

def function(x):
    output = x**3 + x**2 + x + 5
    return output
```

## 2. Integrand Plot

The code below creates the integrand's plot. `integrand(a, b, step=100)` establishes a plot for the integrand defined with `function(x)`. This plot will serve as the basis for the graphs of approximations later discussed.

The function is set like this...

`integrand(a, b, step=100)`:
- a : The integrand's left bound
- b : The integrand's right bound
- step : The number of points defined for the function on $a \leq x \leq b$; the higher the number, the more accurate the plot.

```python
def integrand(a, b, step=100):
    # Axes
    x_axis = np.linspace(a,b,step)
    y_axis = function(x_axis)

    if np.nan in y_axis:
        print('WARNING! There are values outside the domain!')
        pass

    # Setup Plots
    plt.style.use('classic')

    global fig
    global ax

    fig, ax = plt.subplots()

    ax.plot(x_axis, y_axis, color='#000000')

    ax.set_title('Integrand Plot')
    ax.set_xlabel('X-Axis')
    ax.set_ylabel('Y-Axis')

integrand(0, 5, 100)
```

## 3. Left-Hand Riemann Sums

A Left-Hand Riemann Sum estimates a definite integral from x = a to x = b with n rectangular subdivisions that touch the curve with their top-left points. The program first creates n subintervals between x = a and x = b and then raises these subintervals into rectangles. These rectangles then have their areas added together into an integral estimation.

$$\int_a^b f(x) \: \mathrm{d}x 
\approx 
\dfrac{b-a}{n} \times \sum_{k=0}^{n-1} f(a + \dfrac{b-a}{n}k)$$

###### My interpretation of a Left-Hand Riemann Sum formula

To both graph and calculate a Left-Hand Riemann Sum, I used this function containing steps of code.

`LeftHandSum(a = 0, b = 10, n = 10)`:
- a: Defines the left-most part of an indefinite integral's region
- b: Defines the right-most part of an indefinite integral's region
- n: Defines the number of subdivisions between the domain $a \leq x \leq b$

The steps are listed on the code.

```python
def LeftHandSum(a=0, b=10, n=10):
    # Variables Used
    dx = (b-a)/n
    x_axis = np.arange(a, b, dx)
    y_axis = function(x_axis)

    # Calculations
    left_hand_sum = np.sum(y_axis)*dx
    print('Left-Hand Sum: ' + str(left_hand_sum))

    # Plot
    integrand(a, b, 500)
    ax.bar(x_axis + dx/2, y_axis, width = dx, color='#000000', alpha=0.5)
    ax.set_title('Left-Hand Sum Plot')
    plt.show()

LeftHandSum(n=5)
```

## 4. Right-Hand Riemann Sums

Right-Hand Riemann Sums estimate an indefinite integral similarly to Left-Hand Riemann Sums. n rectangular subdivisions with top-right points touching the integrand's curve have their collected areas added together.

$$\int_a^b f(x) \: \mathrm{d}x 
\approx 
\dfrac{b-a}{n} \times \sum_{k=1}^{n} f(a + \dfrac{b-a}{n}k)$$

###### My interpretation of a Right-Hand Riemann Sum formula

Like the Left-Hand Riemann Sum's code, I both graphed and calculated the Right-Hand Riemann Sum's code using the following function with lines of code separated into three steps. 

`RightHandSum(a = 0, b = 10, n = 10)`:
- a: Defines the left-most part of an indefinite integral's region
- b: Defines the right-most part of an indefinite integral's region
- n: Defines the number of subdivisions between the domain $a \leq x \leq b$

The steps are listed on the code.

```python
def RightHandSum(a=0, b=10, n=10):
    # Variables Used
    dx = (b-a)/n
    x_axis = np.arange(a + dx, b + dx, dx)
    y_axis = function(x_axis)

    # Calculations
    right_hand_sum = np.sum(y_axis)*dx
    print('Left-Hand Sum: ' + str(right_hand_sum))

    # Plot
    integrand(a, b, 500)
    ax.bar(x_axis - dx/2, y_axis, width = dx, color='#000000', alpha=0.5)
    ax.set_title('Right-Hand Sum Plot')
    plt.show()

RightHandSum()
```

## 5. Midpoint Riemann Sums

Midpoint Riemann Sums involves defining n rectangular subdivisions between x = a and x = b. These rectangles have heights such that middle of their tops touch the curve and get their areas added together to get a estimation.

$$\int_a^b f(x) \: \mathrm{d}x 
\approx 
\dfrac{b-a}{n} \times \sum_{k=1}^{n} f(a + \dfrac{b-a}{n}(k+\dfrac{1}{2}))$$

###### My interpretation of a Midpoint Riemann Sum formula

Once again, the graphing and calculating of the Midpoint Riemann Sum was done with a function containing code separated into three portions.

This is the function...

`MidPointSum(a = 0, b = 10, n = 10)`:
- a: Defines the left-most part of an indefinite integral's region
- b: Defines the right-most part of an indefinite integral's region
- n: Defines the number of subdivisions between the domain $a \leq x \leq b$

The steps are listed on the code.

```python
def MidPointSum(a=0, b=10, n=10):
    # Variables Used
    dx = (b-a)/n
    x_axis = np.arange(a, b, dx) + dx/2
    y_axis = function(x_axis)

    # Calculations
    mid_point_sum = np.sum(y_axis)*dx
    print('MidPoint Sum Calculation: ' + str(mid_point_sum))

    # Plot
    integrand(a, b)
    ax.bar(x_axis, y_axis, width=dx, color='#000000', alpha=0.5)
    ax.set_title('MidPoint Sum Plot')
    plt.show()

MidPointSum()
```

## 6. Trapezoidal Riemann Sums

Trapezoidal Riemann Sums calculate the area below the integrand's plot with trapezoids rather than rectangles. 

Algebraically, this means that the method would calculate an estimate by averaging the values returned by a Left-Hand Sum and a Right-Hand Sum. Another method involves first summing different function points along n subintervals between x = a and x = b (with the middle points, let say $f(x_{2})$, ..., $f(x_{n-1})$, being multplied by 2 and the left-most and right-most points being multiplied by 1). Then, the sum is divided by a value of 2 and multplied by the size of each subinterval used (i.e., $\dfrac{b-a}{n}$).

$$\int_a^b f(x) \: \mathrm{d}x 
\approx 
\dfrac{b-a}{n} \times [\dfrac{f(x_{1}) + 2f(x_{2}) + 2f(x_{3}) ... 2f(x_{n-1}) + f(x_{n})}{2}]$$

###### My interpretation of a Trapezoidal Riemann Sum

The code below showcases the function and steps of code I used to graph and calculate a Trapezoidal Riemann Sums:

`TrapezoidalSum(a = 0, b = 10, n = 10)`:
- a: Defines the left-most part of an indefinite integral's region
- b: Defines the right-most part of an indefinite integral's region
- n: Defines the number of subdivisions between the domain $a \leq x \leq b$

The setps are listed in the code.

```python
def TrapezoidalSum(a=0, b=10, n=10):
    # Variables Used
    dx = (b-a)/n
    x_axis = np.arange(a, b + dx, dx)
    y_axis = function(x_axis)

    # Calculations
    trapezoidal_sum = (np.sum(y_axis*2) - y_axis[0] - y_axis[-1])*0.5*dx
    print('Trapezoidal Sum: ' + str(trapezoidal_sum))

    # Plot
    integrand(a,b)
    ax.fill_between(x_axis, 0, y_axis, color=(0,0,0,0.5), ec=(0,0,0,1), lw = 1)
    ax.set_title('Trapezoidal Sum Plot')
    plt.show()

TrapezoidalSum(n=5)
```
